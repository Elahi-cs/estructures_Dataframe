* **Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)**

Calcularem primer la velocitat mitjana, en milles per hora.

```python
# En mph
df['AvgSpeed'] = df.Distance / (df.ActualElapsedTime/60)
df['AvgSpeed']
```

A continuació, crearem una columna que ens indicarà si el vol ha arribat tard o no. Farem servir la mateixa convenció que utilitza el dataset en les columnes ``Cancelled`` i ``Diverted``, on ``0`` és ``False`` i ``1`` és ``True``

```python
df['Delayed'] = np.where(df.ArrDelay > 0, 1, 0)
df['Delayed']
```

Ara podem veure com ha quedat el dataframe amb les columnes noves.

```python
df.head()
```

* **Taula de les aerolínies amb més endarreriments acumulats**

Per a aquesta informació podem fer servir la columna de UniqueCarrier. Però, com sabem quina aerolínia es quina? Ampliarem el nostre dataset fent servir dades adicionals, proveïdes per la [font del dataset original](http://stat-computing.org/dataexpo/2009/supplemental-data.html). Concretament, carregarem un arxiu csv amb els codis de cada aerolínia, anomenat ``carriers.csv``.

```python
carriers = pd.read_csv('carriers.csv')
carriers.set_index('Code', inplace=True)
carriers.head()
```

I a continuació l'incorporarem al dataset que estàvem manipulant:

```python
df = pd.merge(df, carriers, how='left', left_on='UniqueCarrier', right_on='Code')
df.rename(columns={'Description': 'Carrier'}, inplace=True)
df.drop('UniqueCarrier', axis=1, inplace=True)
df.head()
```

Un cop tenim les dades necessàries, crearem una taula nova per a accedir a la informació que volem:

```python
delays = df[['Delayed', 'Carrier']].copy()
delayed_carriers = delays.groupby('Carrier').aggregate(sum)
delayed_carriers.sort_values(by='Delayed', axis=0, ascending=False)
```

* **Quins són els vols més llargs? I els més endarrerits?**

De manera similar a l'exercici anterior, carregarem un dataset amb els codis dels aeroports de [la font](http://stat-computing.org/dataexpo/2009/supplemental-data.html) (``airports.csv``), l'integrarem al dataset amb al que estem treballant, i crearem una taula nova només amb les columnes rellevants al que se'ns demana.

```python
airports = pd.read_csv('airports.csv')
airports.set_index('iata', inplace=True)
airports.head()
```

```python
airport_codes = airports[['airport']]
df = df.merge(airport_codes, how='left', left_on='Origin', right_on='iata')
df = df.merge(airport_codes, how='left', left_on='Dest', right_on='iata')
df.rename(columns={'airport_x': 'OriginAirp', 'airport_y': 'DestAirp'}, inplace=True)
df['Flight'] = df['OriginAirp'] + ' - ' + df['DestAirp']
df.head()
```

Amb aquestes dades noves ja podem identificar els vols amb més claredat. A continuació crearem una taula nova i començarem a buscar els vols més llargs:

```python
flights = df[['ActualElapsedTime', 'ArrDelay', 'Flight', 'OriginAirp', 'DestAirp']].copy()
longest_flights = flights.groupby('Flight')
longest_flights.mean().sort_values(by='ActualElapsedTime', axis=0, ascending=False).head()
```

I, finalment, els més endarrerits:

```python
delayed_flights = flights.groupby('Flight')
delayed_flights.mean().sort_values(by='ArrDelay', axis=0, ascending=False).head()
```
